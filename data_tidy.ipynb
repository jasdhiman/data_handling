```python
# let us tidy the data when one of the column has two different types of variables
# e.g. here element has both tmin and tmax
weather = pd.read_csv('C:/Users/jaska/Desktop/tutorial/data/weather.csv')
```

```python
weather.head()
```

```python
weather.melt(id_vars=['id', 'year', 'month', 'element'],
            var_name = 'day',
            value_name = 'temp')
```

```python
weather_long = weather.melt(id_vars=['id', 'year', 'month', 'element'],
            var_name = 'day',
            value_name = 'temp')
```

```python
# using pivot table to separate two variables contained in one column (tmin and tmax)
weather_long.pivot_table(index = ['id', 'year', 'month', 'day'],
                        columns = 'element',
                        values = 'temp')
```

```python
# using reset.index() command to fill missing cells in the table for ease of use
weather_long.pivot_table(index = ['id', 'year', 'month', 'day'],
                        columns = 'element',
                        values = 'temp').reset_index()
```
